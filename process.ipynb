```python
import pandas as pd
```

```python
years = [15,16,17,18,19]

datasets = {}

for year in years:
    df = pd.read_csv(f'./processed_data/{year}.csv')
    df = df.rename(str.lower, axis='columns')
    df = df.rename(columns = {'bldgclass': str(year)})
    datasets[year] = df
```

```python
datasets[15].merge(datasets[16], on = "bbl").merge(datasets[17], on ="bbl").merge(datasets[19], on ="bbl")
```

```python
merged = None
for i, year in enumerate(years):
    if i == 0:
        merged = datasets[year]
    else:
        merged = pd.merge(merged, datasets[year])
```

```python
merged
```

```python
m = merged[merged["15"].str.startswith('M') | merged["16"].str.startswith('M') | merged["17"].str.startswith('M') | merged["18"].str.startswith('M') | merged["19"].str.startswith('M')]
```

```python
pluto = pd.read_csv('./data/19v2.csv')
mn_pluto = pluto[pluto["borough"] == "MN"][["bbl","cd","address","xcoord","ycoord"]].copy()
```

```python
m2 = pd.merge(m, mn_pluto, on = "bbl")
```
