```python
dob_jobs = pd.read_csv('./dob_data/dob_job.csv')
dob_jobs.head()
```

```python
mn_dob_jobs = dob_jobs[dob_jobs['Borough'] == 'MANHATTAN'].copy()
```

```python
def generateBBL(row):
    if not pd.isna(row["Block"]):
        block = int(row["Block"])
        row = str(row["Lot"])[1:]
        return f'1{block:05}{row}'
    else:
        return None

mn_dob_jobs['bbl'] = mn_dob_jobs.apply(generateBBL ,axis = 1)
```
